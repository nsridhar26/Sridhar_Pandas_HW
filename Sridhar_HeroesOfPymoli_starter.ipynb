```python
# Dependencies and Setup
import pandas as pd


# File to Load (Remember to Change These)
file_to_load = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_df = pd.read_csv(file_to_load, encoding="utf-8")
purchase_df.head()
```

## Player Count

* Display the total number of players


```python
# Calculate the Number of Unique Players
player_demographics = purchase_df.loc[:, ['Gender', 'SN', 'Age']]
player_demographics = player_demographics.drop_duplicates()
num_players = player_demographics.count()[0]# Display the total number of players
pd.DataFrame({'Total Players': [num_players]})
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
#unique = purchase_df['Item Name'].unique()
#unique
Items_count = len(purchase_df['Item ID'].unique())
# The value_counts method counts unique values in a column
#count = purchase_df['Item Name'].value_counts()
#count
#calculate avg price, number of purchases & Total Revenue
avg_price = purchase_df['Price'].mean()
number_purchases = purchase_df ['Purchase ID'].count()
total_revenue = purchase_df ['Price'].sum()

#summary table
summary_table=pd.DataFrame({"Number of Unique Items": [Items_count], "Average Price": [avg_price], "Number of Purchases":\
                            [number_purchases], "Total Revenue": [total_revenue]})
#data munging
summary_table['Average Price']= summary_table['Average Price'].map('${:.2f}'.format)
summary_table['Total Revenue']= summary_table['Total Revenue'].map('${:,.2f}'.format)
summary_table
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
#extract data
item_data = purchase_df.loc[:,['Item ID', 'Item Name', 'Price']]

#purchase_df = purchase_df.sort_values('Price', ascending=False)
total_purchase=item_data.groupby(['Item Name', 'Item ID']).sum()['Price'].rename('Total Value')
purchase_count=item_data.groupby(['Item Name', 'Item ID']).count()['Price'].rename ('Purchase Count')
item_price=item_data.groupby(['Item Name', 'Item ID']).mean()['Price']


#organized data frame
most_popular = pd.DataFrame({'Purchase Count': purchase_count, 'Total Value': total_purchase, 'Item Price': item_price})

# example: freedom_df = happiness_df.sort_values("Freedom", ascending=False)
most_popular_sorted =most_popular.sort_values('Purchase Count', axis = 0, ascending=False)

#Minor Data Munging
most_popular_sorted['Item Price'] = most_popular_sorted['Item Price'].map('${:,.2f}'.format)
most_popular_sorted['Purchase Count'] = most_popular_sorted['Purchase Count'].map('{:,}'.format)
most_popular_sorted['Total Purchase Value'] = most_popular_sorted['Total Value'].map('${:,.2f}'.format)
item_popularity = most_popular_sorted.loc[:,['Purchase Count', 'Item Price', 'Total Value']]

item_popularity.head(5)
```
